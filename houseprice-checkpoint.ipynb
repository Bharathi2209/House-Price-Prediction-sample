```python
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
house_price_dataset = pd.read_csv('C:\\Users\\Bharathi\\Desktop\\ml\\.ipynb_checkpoints\\House Price Prediction Dataset.csv')

```

```python
print(house_price_dataset)
```

```python
house_price_dataframe=pd.DataFrame(house_price_dataset)
```

```python
house_price_dataframe.head()
```

```python
house_price_dataframe['Price']=house_price_dataset['Price']
```

```python
house_price_dataframe.shape
```

```python
house_price_dataframe.isnull().sum()
```

```python
house_price_dataframe.describe()
```

```python
# Convert 'Garage' column to numeric values using one-hot encoding
house_price_dataframe = pd.get_dummies(house_price_dataframe, columns=['Garage'])

# Calculate the correlation
correlation = house_price_dataframe.corr()
print(correlation)
```

```python
#constructing map for understanding corelation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt=".1f",annot=True,annot_kws={"size":8},cmap="Blues")
```

```python
#splitting data and price

X=house_price_dataframe.drop(["Price"],axis=1)
Y=house_price_dataframe["Price"]
```

```python
print(X)
print(Y)
```

```python
#splitting data into train data and test data

X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#model training -- xgboost regressor model

#type of decision tree model
model = XGBRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)
```

```python
model.fit(X_train,Y_train)
```

```python
predictions=model.predict(X_train)

print(predictions)
```

```python
predictions=model.predict(X_test)

print(predictions)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(Y_train, predictions)
print(f'Mean Squared Error: {mse}')
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(Y_test, predictions)
print(f'Mean Squared Error: {mse}')
```

```python
r2 = r2_score(Y_train, predictions)
print(f'R-squared Score: {r2}')
```

```python
r2 = r2_score(Y_test, predictions)
print(f'R-squared Score: {r2}')
```

```python
#visualizing the actual prices and predicted prices

plt.scatter(Y_train, predictions)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Vs Predicted")
plt.show()
```
